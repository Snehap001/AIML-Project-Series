```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('training.csv')
X = dataset.drop('prognosis', axis=1)
y = dataset['prognosis'] 
```

```python
print(y)
```

```python
from sklearn.preprocessing import LabelEncoder

# Creating an instance of LabelEncoder
label_encoder = LabelEncoder()

# Fitting and transforming thoutput variable to convert categorical data to numerical labels
y_encoded = label_encoder.fit_transform(y)
```

```python
from sklearn.preprocessing import StandardScaler

# Creating an instance of StandardScaler
sc_X = StandardScaler()

# Fitting and transforming the input features
X_scaled = sc_X.fit_transform(X)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_scaled, y_encoded)
```

```python
import pandas as pd

# Load the testing dataset
test_data = pd.read_csv('Testing.csv')

# Assuming the output variable 'y' and input features 'X' are in the test dataset
X_test = test_data.drop('prognosis', axis=1)  # replace 'disease' with the actual column name for the output
y_test = test_data['prognosis']  #
print("Training columns:", dataset.columns)
print("Testing columns:", test_data.columns)
assert X.shape[1] == X_test.shape[1], "Feature count mismatch between training and testing datasets"
```

```python
y_test_encoded = label_encoder.transform(y_test)

# Scale the input features of the test data
X_test_scaled = sc_X.transform(X_test)
```

```python
y_pred_encoded = classifier.predict(X_test_scaled)

# Optionally, convert encoded predictions back to original labels
y_pred = label_encoder.inverse_transform(y_pred_encoded)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Calculate accuracy
accuracy = accuracy_score(y_test_encoded, y_pred_encoded)
print(f'Accuracy: {accuracy:.2f}')

# Generate a classification report
print(classification_report(y_test_encoded, y_pred_encoded, target_names=label_encoder.classes_))

# Generate and display the confusion matrix
cm = confusion_matrix(y_test_encoded, y_pred_encoded)
print(cm)
```
